```python
import h5py
import matplotlib.pyplot as plt
import numpy as np

from src import DATA_DIR
```

```python
def softmax(logits, axis=1):
    logits -= np.max(logits, axis=-1, keepdims=True)
    return np.exp(logits) / np.sum(np.exp(logits), axis=-1, keepdims=True) 
```

```python
def print_weights(model_name):
    with h5py.File("/home/z1079621/storage/embeddings/results/%s/best_model.h5" % model_name) as f:
        for name in ['weight_elmo_embeddings_1', 'weight_elmo_embeddings_2']:
            elmo_weights = f['model_weights'][name][name]

            print(name, "gamma", list(elmo_weights['ELMo_gamma:0']))
#             print(list(elmo_weights.keys()))
#             print("weights", list(elmo_weights['ELMo_W:0']))
            normalized_weights = softmax(elmo_weights['ELMo_W:0'])
#             print("softmax-normalized weights", )    
            plt.bar(x=range(3), height=normalized_weights)
            plt.xticks(range(3))
            plt.show()
```
